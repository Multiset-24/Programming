```python
# python calculator

def add(x,y):
    return x+y
def sub(x,y):
    return x-y
def mul(x,y):
    return x*y
def div(x,y):
    return x/y#divide in python gives float value , for getting integer value use // instead of / which gives floor division not seiling division
def power(x,y):
    return x**y

operation=input("Enter the operation you want to perform: ")
x=int(input("Enter the first number: "))
y=int(input("Enter the second number: "))

if operation=='add':
    print(f"The sum of {x} and {y} is {add(x,y)}")
elif operation=='sub':
    print(f"The difference of {x} and {y} is {sub(x,y)}")
elif operation=='mul':
    print(f"The product of {x} and {y} is {mul(x,y)}")
elif operation=='div':
    print(f"The division of {x} and {y} is {div(x,y)}")
elif operation=='power':
    print(f"The power of {x} raised to {y} is {power(x,y)}")
else:
    print("Invalid operation")
```

```python
user_name=input("Enter your name: ")

def validator_for_name(user_name):
    if len(user_name)>12:#to check the length of the string use len() function
        print("The length of the name is greater than 12")
        return
    elif user_name.find(" ")!=-1:
        print("The name contains space")
        return
    elif not user_name.isalpha():
        print("The name contains digits")
        return
    else:
        print("The name is valid")
        return

validator_for_name(user_name)
```

```python
# countdown timer in days , minutes , seconds

import time

Countdown=int(input("Enter the time in seconds"))

for t in range(Countdown,0,-1):
    seconds=t%60
    minutes=int((t/60)%60)
    hours=int(t/3600)
    print(f"{hours:02}:{minutes:02}:{seconds:02}")

    time.sleep(1)

print("Time's Up!!")


```

```python
#This is quiz game exercise

Questions=["What is the closest planet to sun","Name of the capital of idea","Approx value of pie","SI Unit of length"]
Options=[["Mercury","Venus","Earth","Mars"],
         ["Delhi","Kolkata","Lucknow","Agra"],
         ["22/7","4","2.79","9"],
         ["Meter","Second","Gauss","Meter^2"]]
Answers=[0,0,0,0]
score=0
print("----Welcome----")

for i in range(len(Questions)):
    print(Questions[i])
    for idx,opt in enumerate(Options[i],1):
        print(f"{idx}.{opt}")
    
    print("Enter your answer number")
    answer_given=int(input())

    try:
        if answer_given-1==Answers[i]:
            print("Correct")
            score+=1
        else:
            print("Oops Incorrect")
    except(ValueError,IndexError):
        print("Invalid input")

print("----The End----")
print(f"Your score is {(score/len(Questions))*100}")
print("----Thank you for playing----")   



```

```python
# this is rock paper scissors game

import random 
choices=["Paper","Rock","Scissors"]




user_score=0
computer_score=0

user_choice="Rock"

while(True):
    # cc=random.randint(0,len(choices)-1) # using indices
    computer_choice=random.choice(choices) #direct way 
    user_choice=input("Pick option").capitalize()

    if user_choice=="Exit":
        break
    elif user_choice.isdigit() or (user_choice != "Paper" and user_choice != "Rock" and user_choice != "Scissors"): 
        print("Please input a valid choice")
    elif (user_choice=="Rock" and computer_choice=="Scissors") or (user_choice=="Scissors" and computer_choice=="Paper") or (user_choice=="Paper" and computer_choice=="Rock"):
        print("You won")
        user_score+=1
    elif computer_choice==user_choice:
        print("It's a tie Would you like to go on ?")
    else:
        print("I won let's try again ?")
        computer_score+=1

print("Thank you for playing with me")

print(f"Results")

print(f"Your score is {user_score} and mine score is {computer_score}")

if computer_score>=user_score:
    print("Oops Your loose Would you like to try again !!")
else :
    print("Wallah you won !!")


```

```python
# countdown timer--> default args


import time

def count(end,start=0):# default args should succeed the non default args
    for i in range(end,start-1,-1):
        print(i)
        time.sleep(1)
    print("Time's Up!!")

count(6)
```

```python
# Arbitirary arguments && iterables-> any object whose elements can be accessed one by one

# there are 2 types of arbitirary arguments , args-tuples in function and kwargs-dictionary in function

def print_sum(*nums):
    sum=0
    for x in nums:
        sum+=x
    print(sum)

print_sum(1,2)
print_sum(10,2,9,7,100)

def hirarchy(*th_level,**kwargs,):
    for key,val in kwargs.items():#dictionary is an iterable
        print(f"{key}:{val}")
    
    print(th_level)

hirarchy(
    12,12,11,13,11,12,
    leader="Shiveet",
    co_leader="Parth, saksham , bhargav",
    elder="Saurav,amit"
)
```

```python
# this is going to be alarm clock

import time
import pygame
import datetime

audio_file="Cock_sound_for_alarm.mp3"

if __name__=="__main__":
    Alarm_time=input("Exter the Alarm time in HH:MM:SS format")

    def alarm_clock(Alarm_time):
        while True:
            curr_time=datetime.datetime.now().strftime("%H:%M:%S")

            if curr_time==Alarm_time:
                print("Wake up Bro It's your exam !!")
                pygame.mixer.init()
                pygame.mixer.music.load(audio_file)
                pygame.mixer.music.play()

                while pygame.mixer.music.get_busy():
                    user_input=input("Enter S to stop the clock")
                    if user_input.capitalize()=="S":
                        print("Good morning")
                        pygame.mixer.music.stop()
                    time.sleep(1)

                break
            
            print(curr_time)
            time.sleep(1)

    alarm_clock(Alarm_time)


        

```

```python
#this code gets pokemon data 

import requests

base_url="https://pokeapi.co/api/v2/"

pokemon_name=input("Enter the name of the pokemon you want info about")


url=f"{base_url}/pokemon/{pokemon_name}"

data=requests.get(url)

if data:
    if data.status_code==200:
        print("Data retrieved successfully !!")
        info=dict(data.json())
        print(info,end=" ")
        for key,val in info.items():
            print(key,val)
            
else:
    print("Something went wrong !!")
```
